De maneira geral, uma análise de dados tem os seguintes passos:
- **Coleta de Dados**: Muito possivelmente você vai se deparar com algum problema ao longo da sua vida em que você precisará de dados, os quais não necessariamente existem de maneira fácil. Logo, você precisará coletar esses dados, e uma maneira muito utilizada para a coleta de dados da Web é o chamado Web Scraping. 

- **Preprocessamento dos Dados**: Infelizmente, tanto dados coletados quanto dados disponibilizados não são perfeitos. Você pode encontrar uma série de problemas como espaços vazios, dados inconsistentes, variáveis irrelevantes, problemas de leitura dos dados, etc. Saber lidar com esses problemas e deixar sua base de dados pronta para uma análise é essencial.

- **Análise de Dados**: Aqui é quando a magia acontece. Nesse passo é que você vai retirar insights dos dados, identificar padrões, fazer previsões, etc. É importante saber como lidar com cada tipo de análise, pois é essencialmente aqui que suas perguntas vão ser respondidas. Esperamos que corretamente.

- **Visualização dos Dados**: Muitas vezes é necessário apresentar os resultados obtidos de forma visual, muitas vezes até interativa, mas muitas vezes até mesmo num csv. Saber como apresentar seu trabalho para algum público pode ser crucial.

## 06. Classes e Métodos

```python
#vamos criar um novo Type, que chamaremos de circulo
#como todo circulo, ele tem q ter algumas propriedades: raio, area, comprimento

pi = 3.14
```

```python
class circle:
    
    def __init__(self, radius):
        self.radius = radius
    
    def area(self):
        return pi * self.radius ** 2
    
    def circumference(self):
        return 2 * pi * self.radius
```

```python
c = circle(10)
```

```python
c.radius
```

```python
c.area()
```

```python
c.circumference()
```

```python
type(c.area)
```

```python
type(c.area())
```

```python
type(c)
```
